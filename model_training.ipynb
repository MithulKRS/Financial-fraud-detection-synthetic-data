```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
```

```python
x_train=pd.read_csv('Artifacts/xtrain_data.csv')
x_test=pd.read_csv('Artifacts/xtesting_data.csv')
y_train=pd.read_csv('Artifacts/ytrain_data.csv')
y_test=pd.read_csv('Artifacts/ytest_data.csv')
y_train.shape
y_test.shape
```

```python
x_train=x_train.drop(['Unnamed: 0','remainder__is_fraud'],axis=1)
x_test=x_test.drop(['Unnamed: 0','remainder__is_fraud'],axis=1)
y_test=y_test.drop('Unnamed: 0',axis=1)
print(y_test.shape)
```

```python
y_train=y_train.drop('Unnamed: 0',axis=1)
```

```python
x_train
```

```python
model = LogisticRegression(max_iter=1000,penalty='l2')
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score, confusion_matrix
```

```python
y_pred.shape
```

```python
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}\n")
```

```python
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(cm, "\n")

# --- 3. Extract TP, TN, FP, FN ---
# The ravel() function safely unpacks the 2x2 matrix into four separate variables
tn, fp, fn, tp = cm.ravel()

print(f"True Positives (TP): {tp} (Actual fraud caught by the model)")
print(f"False Positives (FP): {fp} (Legitimate transactions falsely flagged as fraud)")
print(f"True Negatives (TN): {tn} (Legitimate transactions correctly identified)")
print(f"False Negatives (FN): {fn} (Actual fraud missed by the model)")
```
