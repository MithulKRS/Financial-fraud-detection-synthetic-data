```python
import shap
import pandas as pd
import matplotlib.pyplot as plt
import joblib
```

```python
x_train=pd.read_csv("Artifacts/xtrain_data.csv")
model =joblib.load('model.joblib')
```

```python
x_train=x_train.drop(['Unnamed: 0','remainder__is_fraud'],axis=1)
```

```python
explainer = shap.LinearExplainer(model, x_train)
```

```python
x_test=pd.read_csv('Artifacts/xtesting_data.csv')
```

```python
x_test=x_test.drop(['Unnamed: 0','remainder__is_fraud'],axis=1)
```

```python
shap_values = explainer(x_test)
```

```python
plt.figure(figsize=(10, 6))
shap.summary_plot(shap_values, x_test, plot_type="bar")
```

```python
transaction_index = 1996
```

```python
print(f"\nGenerating Local Explanation for Transaction {transaction_index} (Waterfall Plot)...")
# Waterfall plots show how each feature pushed the model's output from the "base value" to the final prediction
shap.plots.waterfall(shap_values[transaction_index])
```
